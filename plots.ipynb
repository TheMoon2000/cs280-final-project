```python
import numpy as np
import matplotlib.pyplot as plt
from typing import List
import pickle
```

```python
def graph_stat(model_names: List[str], stat: str, max_epochs=150, smooth_size=1):

    plt.figure(facecolor='white', figsize=(12, 5))
    plt.title(f'{len(model_names)} models on {stat}')
    plt.xlabel('Epochs')
    plt.ylabel(stat)
    for name in model_names:
        with open(f'models/{name}/stats.pk', 'rb') as f:
            data = pickle.load(f)
        epochs = range(len(data[:max_epochs]))
        stat_values = [x[stat] for x in data[:max_epochs]]
        plt.plot(epochs, stat_values, label=name, alpha=0.6)
        plt.legend()
    
#     plt.xticks(epochs)
```

```python
graph_stat(['no_dropout_8', 'dropout_0.1', 'dropout_0.2'], 'eval_psnr', max_epochs=1000)
```
